### 1. Известно, что генеральная совокупность распределена нормально со средним квадратическим отклонением, равным 16. Найти доверительный интервал для оценки математического ожидания a с надежностью 0.95, если выборочная средняя M = 80, а объем выборки n = 256.

$$\bar{x} \pm z_{\frac{\alpha}{2}} * \frac{\sigma}{\sqrt{n}}$$

```python
s = 16
x = 80
n = 256

z_table = stats.norm.ppf(0.975) # надежность 0.95 => alpha = 0.05 => односторонний интервал = 0.975

x - z_table * s / np.sqrt(n), x + z_table * s / np.sqrt(n)
```

### 2. В результате 10 независимых измерений некоторой величины X, выполненных с одинаковой точностью, получены опытные данные: 6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1 Предполагая, что результаты измерений подчинены нормальному закону распределения вероятностей, оценить истинное значение величины X при помощи доверительного интервала, покрывающего это значение с доверительной вероятностью 0,95.

$$\bar{x} \pm t_{\frac{\alpha}{2}} * \frac{\sigma}{\sqrt{n}}$$

```python
x = np.array([6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1 ])
x_mean = np.mean(x)
x_n = len(x)

D = np.var(x, ddof = 1) # несмещенная дисперсия генеральной совокупности для нормального распределения
s = np.std(x, ddof = 1) # несмещенное среднее квадратное отклонение для нормального распределения
t_table = stats.t.ppf(0.975, x_n - 1) # t критерий для 95% интервала

print(f'Значение величины Х лежит в интервале {x_mean - t_table * s / np.sqrt(x_n), x_mean + t_table * s / np.sqrt(x_n)}')
```

$$\Delta \pm t_{\frac{\alpha}{2}} * S_\Delta;
\Delta = \bar{x}_1 - \bar{x}_2;
S_\Delta = \sqrt{\frac{D}{n_1}+\frac{D}{n_2}};
D = \frac{1}{2}(D_1+D_2);$$

```python
height_daughter = d = [175, 167, 154, 174, 178, 148, 160, 167, 169, 170]
height_mother = m = [178, 165, 165, 173, 168, 155, 160, 164, 178, 175]

n_d = len(d)
n_m = len(m)

d_mean = np.mean(d)
m_mean = np.mean(m)
delta = d_mean - m_mean

D_d = np.var(d, ddof = 1)
D_m = np.var(m, ddof = 1)
D = 1/2 * (D_d + D_m)
se = np.sqrt(D / n_d + D / n_m) # стандартная ошибка среднего для разности

t = stats.t.ppf(0.975, n_d + n_m - 2) # t критерий для 95% интервала, 2 выборки 

L = delta - t * se #
U = delta + t * se #
print(f'Интервал для разности среднего роста родителей и детей {L, U}')
```

1. На препарате А положительный результат лечения наблюдается у 17 из 32 пациентов, а 
на препарате В у 9 из 22. Построить 95% доверительный интервал для разности долей.
Обнаружены ли статистически значимые различия?

```python
m1 = 17
m2 = 9
n1 = 32
n2 = 22
p1 = m1 / n1
p2 = m2 / n2
delta = p1 - p2 # разность долей

z_table = stats.norm.ppf(0.975) # нахождение z критерия. norm для критерия z, t для стьюдента
p = (m1 + m2) / (n1 + n2)
se = (p * (1 - p) * (1 / n1 + 1 / n2)) ** 0.5

delta - z_table * se, delta + z_table * se # границы разностей
# Так как граница разности включает 0, то статистически значимых различий не обнаружено.
# delta проходит через 0. разность не имеет различий. p1 - p2 = delta = 0 => p1 = p2
```

4. С помощью 90% доверительного интервала оценить средний вес нормально 
распределенной популяции, если дисперсия генеральной совокупности 3.6, а среднее 
mарифметичекое по выборке объемом 100 получилось равным 71.2.

```python
x = 71.2
n = 100
D = 3.6
se = (D / n) ** 0.5

z_table = stats.norm.ppf(0.95)
# с вероятностью 90% данный интервал хранит истинное значения веса генеральной совокупности
x - se * z_table, x + se * z_table
```

5. Найдите 95% доверительные интервалы для долей больных, которые не чувствовали боли 
при включенном и выключенном приборе. Можно ли по этим интервалам оценить 
статистическую значимость различий?

```python
# Прибор включен: Боли нет - 24, боль есть 6
# Прибор выключен: Боли нет - 3, боль есть 17

# n * p = 30 * 24/30 = 24  => n * p > 5
# n * (1 - p) = 30 * (1 - 24/30) = 6 => n * (1 - p) > 5
# !!! n * p = 20 * 3/20 = 3 => n * p < 5
# n * (1 - p) = 20 * (1 - 3/20) = 17 => n * (1 - p) > 5

# если условие не выполняется, то используем биноминально распределение (формула Бернули)

# 2 интервал (0.03; 0.38) по графику => пересечение выборочной доли (3/20 = 0.15) и размеров выборки 20
# 1 интервал:
m1 = 24
n1 = 30
p1 = m1 / n1
se = (p1 * (1 - p1) / n1) ** 0.5
z_t = stats.norm.ppf(0.975)
p1 - se * z_t, p1 + se * z_t

# прибор эффективен, статистическая значимость > чем вторая доля
```

6. Даны две группы студентов. В каждой по 21 человеку. Первая группа прошла недельную 
подготовку для сдачи тестов, другая группа приступила к тестам без подготовки.
В 1й группе среднее время прохождения теста целиком требует 1.8 ч, во второй 2.3 ч. 
Постройте 95% интервал для разности средних арифметических, если известно, что 
стандартное отклонение 1 группы составило 0.4 ч, во 2й группе -0.6ч

```python
x1 = 1.8
x2 = 2.3
delta = x1 - x2
s1 = 0.4
s2 = 0.6

D1 = s1 ** 2
D2 = s2 ** 2
D = (D1 + D2) / 2
se = (D / n1 + D2 / n2) ** 0.5
t_table = stats.t.ppf(0.975, n1 + n2 - 2)

delta - t_table * se, delta + t_table * se
# не являются статистически значимыми, delta < 0 => x1 < x2
# первая группа потратила времени статистически меньше чем вторая. недельная подготовка повлияла на сдачу тестов
```
